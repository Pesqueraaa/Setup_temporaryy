```python
import hashlib, os

nombre_archivo = "00_setup_PesqueraDavid.ipynb"

# Si el archivo no existe en el Colab aún creamos un marcador temporal para poder generar el hash
if not os.path.exists(nombre_archivo):
    with open(nombre_archivo, "w") as f:
        f.write("Plantilla reproducible - archivo inicial")

with open(nombre_archivo, "rb") as f:
    contenido = f.read()

sha = hashlib.sha256(contenido).hexdigest()
print("SHA-256 generado:\n", sha)
```

```python
import json, datetime

registro = {
    "archivo": nombre_archivo,
    "sha256": sha,
    "fecha": datetime.datetime.now().isoformat()
}

with open("registro_hash.json", "w") as f:
    json.dump(registro, f, indent=2)

print("Registro creado: registro_hash.json ✅")
```
